```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection='3d')

theta=np.linspace(0,2*np.pi, 100)
r=1
x=r*np.cos(theta)
y=r*np.sin(theta)
z=np.zeros_like(theta)

ax.plot(x,y,z, 'b-', linewidth=2,
label='Lingkaran di XY plane')

x2=r*np.cos(theta)
y2=np.zeros_like(theta)
z2=r*np.sin(theta)
ax.plot(x2,y2,z2,'r-', linewidth=2,
label='Lingkaran di XZ plane')
ax.set_label('Y')

x3=np.zeros_like(theta)
y3=r*np.cos(theta)
z3=r*np.sin(theta)
ax.plot(x3,y3,z3,'g-', linewidth=2,
label='Lingkaran di YZ plane')
ax.set_xlabel('x')
ax.set_zlim(-1, 1)
ax.set_title('Lingkaran dalam 3 Dimensi')
ax.legend()
plt.show()
```
