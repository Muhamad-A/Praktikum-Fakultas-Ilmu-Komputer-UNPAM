```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('/Users/irvimorantt/Downloads/awannn.jpg') 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel_size = (5, 5)
mean_filtered = cv2.blur(img_rgb, kernel_size)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Citra Asli")
plt.axis('on')

plt.subplot(1, 2, 2)
plt.imshow(mean_filtered)
plt.title(f'Hasil Mean Filter {kernel_size}')
plt.axis('off')

plt.tight_layout()
plt.show()
```
