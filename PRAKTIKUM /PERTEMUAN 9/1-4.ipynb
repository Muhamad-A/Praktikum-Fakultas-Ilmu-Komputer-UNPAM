```python
import cv2
import matplotlib.pyplot as plt

# 1. Membaca gambar dalam mode Grayscale (skala keabuan)
# Ganti 'path_gambar_kamu.jpg' dengan nama atau lokasi file gambar asli Anda
gambar = cv2.imread('path_gambar_kamu.jpg', cv2.IMREAD_GRAYSCALE)

# Pastikan gambar berhasil dimuat sebelum menghitung histogram
if gambar is None:
    print("Error: File gambar tidak ditemukan. Periksa kembali nama file atau letak folder Anda.")
else:
    # 2. Menghitung histogram citra
    histogram = cv2.calcHist([gambar], [0], None, [256], [0, 256])

    # 3. Membuat visualisasi plot
    plt.figure() # Membuat jendela grafik baru
    plt.title("Gambar Histogram")
    plt.xlabel("Intensitas Pixel")  # <-- Menambahkan label sumbu X
    plt.ylabel("Jumlah pixel")     # <-- Menambahkan label sumbu Y
    
    plt.plot(histogram)
    plt.xlim(0, 256)
    plt.show()
```
