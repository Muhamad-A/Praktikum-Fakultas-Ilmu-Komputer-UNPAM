```python
import cv2
import matplotlib.pyplot as plt

gambar_rgb = plt.imread('/Users/irvimorantt/Downloads/awannn.jpg')
gambar = cv2.cvtColor(gambar_rgb, cv2.COLOR_RGB2GRAY)
histogram = cv2.calcHist([gambar], [0], None, [256], [0, 256])

plt.figure(figsize=(8, 5))
plt.title("Gambar Histogram")
plt.ylabel("Jumlah pixel")
plt.xlabel("Intensitas Pixel")

plt.plot(histogram)
plt.xlim(0, 256)
plt.show()
```
