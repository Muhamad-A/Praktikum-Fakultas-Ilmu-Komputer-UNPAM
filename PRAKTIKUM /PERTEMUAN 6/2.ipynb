```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
rect=patches.Rectangle((0.1, 0.6), 0.2, 0.15, linewidth=2, 
edgecolor='black', facecolor='red')
ax.add_patch(rect)

circle=patches.Circle((0.5, 0.65,), 0.1,
linewidth=2, edgecolor='black', facecolor='blue')
ax.add_patch(circle)

triangle=patches.Polygon([[0.75, 0.6],[0.85, 0.8], [0.65, 0.8]],
linewidth=2, edgecolor='black', facecolor='green')
ax.add_patch(triangle)

ellipse=patches.Ellipse((0.3, 0.25), 0.2, 0.1, angle=45,
linewidth=2, edgecolor='black', facecolor='orange')
ax.add_patch(ellipse)

ax.set_aspect('equal')
ax.axis('on')

plt.title('Empat Bentuk dengan Warna Berbeda')
plt.show()


```
