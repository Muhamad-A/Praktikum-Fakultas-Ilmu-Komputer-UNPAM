```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

fig,ax = plt.subplots()

triangle = patches.Polygon([[1, 1], [2, 3], [3, 1]], closed=True, color='blue')
ax.add_patch(triangle)
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Segitiga Berwarna Biru")
plt.show()
```
