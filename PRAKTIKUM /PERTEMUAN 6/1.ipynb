```python
def rgb_to_hsvr(r, g, b):
    r, g, b = r/255.0, g/255.0, b/255.0
    
    mx = max(r, g, b)
    mn = min(r, g, b)
    df = mx - mn

    if df == 0:
        h = 0
    elif mx == r:
        h = (60 * ((g - b) / df) + 360) % 360
    elif mx == g:
        h = (60 * ((b - r) / df) + 120) % 360
    elif mx == b:
        h = (60 * ((r - g) / df) + 240) % 360

    s=0 if mx== 0 else (df/mx)
    v=mx

    return h, s, v

red   = rgb_to_hsv(255,0,0) 
green = rgb_to_hsv(0,255,0)
blue  = rgb_to_hsv(0,0,255)
white = rgb_to_hsv(255,255,255)
black = rgb_to_hsv(0,0,0) 

print("RGB(255,0,0)->HSV:", red)
print("RGB(0,255,0)->HSV:", green)
print("RGB(0,0,255)->HSV:", blue)
print("RGB(255,255,255)->HSV:", white)
print("RGB(0,0,0)->HSV:", black)
```
