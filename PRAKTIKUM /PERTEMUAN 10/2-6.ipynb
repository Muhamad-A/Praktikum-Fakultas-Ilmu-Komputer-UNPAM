```python
import cv2
import matplotlib.pyplot as plt
path_gambar = '/Users/irvimorantt/Downloads/logo w4rjok.png'
img = cv2.imread(path_gambar)

if img is None:
    print("Waduh, gambarnya belum ketemu!")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_fixed = cv2.resize(img_rgb, (100, 50))
    img_scaled_ratio = cv2.resize(img_rgb, (0, 0), fx=2, fy=2)
    print(f"Ukuran Rasio (2x Lipat) : {img_scaled_ratio.shape[:2]}")

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(img_rgb)
    plt.title('Citra Asli')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(img_fixed)
    plt.title('Fixed Size (100x50)')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(img_scaled_ratio)
    plt.title('Skala Rasio 2x')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
