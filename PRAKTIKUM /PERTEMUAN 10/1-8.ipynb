```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path_gambar = '/Users/irvimorantt/Downloads/logo w4rjok.png'
img = cv2.imread(path_gambar)

if img is None:
    print("Waduh, gambarnya belum ketemu!")
else:
    img_resized = cv2.resize(img, (730, 450))

    img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
    img_inverted = cv2.bitwise_not(img_gray)

    kernel = np.ones((30, 30), dtype=np.uint8)

    img_hasil = cv2.dilate(img_inverted, kernel, iterations=1)

    plt.figure(figsize=(14, 6))

    ax1 = plt.subplot(1, 2, 1)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Gambar Asli')
    ax1.tick_params(labelbottom=True, labelleft=True)
    ax2 = plt.subplot(1, 2, 2)
    plt.imshow(img_hasil, cmap='gray')
    plt.title('Hasil Dilatasi')
    ax2.tick_params(labelbottom=True, labelleft=True)

    plt.tight_layout()
    plt.show()
```
