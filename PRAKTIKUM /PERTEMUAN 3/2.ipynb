```python
import matplotlib.pyplot as plt # Diperbaiki agar standar

def brensenham_line(x1, y1, x2, y2):
    x, y = x1, y1 # Inisialisasi x dan y
    points_x = [x]
    points_y = [y]

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1

    # Indentasi: Kode di bawah ini harus masuk ke dalam fungsi
    if dx > dy:
        p = 2 * dy - dx
        for i in range(dx): # Mulai dari langkah pertama
            if p < 0:
                p += 2 * dy
            else:
                y += sy
                p += 2 * (dy - dx)
            x += sx
            points_x.append(x)
            points_y.append(y)
    else:
        p = 2 * dx - dy
        for i in range(dy):
            if p < 0:
                p += 2 * dx
            else:
                x += sx
                p += 2 * (dx - dy)
            y += sy
            points_x.append(x)
            points_y.append(y)

    return points_x, points_y

# Inisialisasi koordinat
x1, y1 = 2, 3
x2, y2 = 10, 8

x_points, y_points = brensenham_line(x1, y1, x2, y2)

print("ALGORITMA BRENSENHAM")
print("="*50)
print(f'Garis dari ({x1}, {y1}) ke ({x2}, {y2})')
print("-" * 50)
for i in range(len(x_points)):
    print(f"({x_points[i]}, {y_points[i]})")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Diperbaiki: markersize (bukan markers)
plt.plot(x_points, y_points, 'ro', linewidth=2, markersize=8, label='Pixel')
plt.grid(True, alpha=0.3)
plt.title(f'Algoritma Brensenham\nGaris dari ({x1},{y1}) ke ({x2},{y2})')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.axhline(y=0, color='k', linestyle='-')
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Diperbaiki: Tutup kurung yang kurang di f-string Awal
plt.annotate(f'Awal({x1},{y1})', (x1, y1), xytext=(5, 5),
             textcoords='offset points', fontsize=10, fontweight='bold')
plt.annotate(f'Akhir({x2},{y2})', (x2, y2), xytext=(5, 5),
             textcoords='offset points', fontsize=10, fontweight='bold')

plt.xlim(min(x_points)-1, max(x_points)+1)
plt.ylim(min(y_points)-1, max(y_points)+1)
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x_points, y_points, c='red', s=200, marker='s', alpha=0.7, label='Pixel terpilih')
plt.grid(True, alpha=0.5, linestyle='-', linewidth=0.5)

for i in range(min(x_points)-1, max(x_points)+2):
    plt.axvline(x=i, color='gray', linestyle='-')
for i in range(min(y_points)-1, max(y_points)+2):
    # Diperbaiki: color='gray' (tambah tanda =)
    plt.axhline(y=i, color='gray', linestyle='-')

plt.title('Visualisasi Pixel Grid')
plt.xlabel('Sumbu X') # Sebelumnya typo Sumbu Y
plt.ylabel('Sumbu Y') # Sebelumnya typo Sumbu x
# Diperbaiki: x_points (sebelumnya x_poitns)
plt.xlim(min(x_points)-1.5, max(x_points)+1.5)
plt.ylim(min(y_points)-1.5, max(y_points)+1.5)
plt.gca().set_aspect('equal')
plt.legend()
plt.tight_layout()
plt.show()
```
