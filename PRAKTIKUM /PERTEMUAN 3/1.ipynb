```python
import matplotlib.pyplot as plt
import numpy as np

def dda_line(x1,y1,x2, y2):
    dx=x2-x1
    dy=y2-y1
    
    steps= int(max(abs(dx), abs(dy)))
    
    x_inc = dx / steps
    y_inc= dy / steps

    x=x1
    y=y1
    
    point_x=[x]
    point_y=[y]

    for _ in range(steps):
        x += x_inc
        y += y_inc
        point_x.append(round(x))
        point_y.append(round(y))
   
    return point_x, points_y

    x1, y1 =2, 3
    x2, y2 =10, 8

    x_points, y_points = dda_line(x1, y1, x2, y2)

    print("Titik-titik garis yang dihasilkan:")
    for i in range(len(x_points)):
        print(f"({x_points[i]}, {y_points[i]})")

plt.figure(figsize=(8, 6))
plt.plot(x_points, y_points,'b-o', linewidth=2, markersize=8)
plt.grid(True, alpha=0.3)
plt.title('Algoritma DDA - Garis dari (2,3) ke (10,8)')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

plt.annotate(f'Awal ({x1},{y1})', (x1,y1), xytext=(5, 5), textcoords='offset points')
plt.annotate(f'Akhir ({x2},{y2})', (x2, y2), xytext=(5, 5), textcoords='offset points')

plt.xlim(min(x_points)-1, max(x_points)+1)
plt.ylim(min(y_points)-1, max(y_points)+1)
plt.show()

```
