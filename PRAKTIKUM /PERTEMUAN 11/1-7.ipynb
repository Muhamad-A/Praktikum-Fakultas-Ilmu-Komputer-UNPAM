```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 500)
y_dengan_aa = np.sin(x)
y_tanpa_aa = y_dengan_aa + 0.5

plt.figure(figsize=(10, 4))
plt.plot(x, y_dengan_aa, label='Dengan AA', linewidth=3, color='#1f77b4')
plt.plot(x, y_tanpa_aa, label='Tanpa AA', linewidth=4, color='#ff7f0e')
plt.title('Perbandingan Antialiasing', fontsize=14, pad=10)
plt.xlim(-0.5, 10.5)
plt.ylim(-1.1, 1.6)
plt.xticks([0, 2, 4, 6, 8, 10])
plt.yticks([-1.0, -0.5, 0.0, 0.5, 1.0, 1.5])
plt.grid(True, linestyle='-', linewidth=0.5, color='#e0e0e0')
plt.legend(loc='lower left', frameon=True, facecolor='white', edgecolor='#e0e0e0')
plt.tight_layout()
plt.show()
```
