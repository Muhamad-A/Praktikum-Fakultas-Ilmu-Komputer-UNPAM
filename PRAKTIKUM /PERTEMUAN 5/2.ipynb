```python
import matplotlib.pyplot as plt
def gambar_lingkaran_midpoint(r):

    titik_x = []
    titik_y = []

    x=0
    y=r
    p=1-r
    
    while x<=y:
        titik_x.extend([x, -x,  x, -x,  y, -y,  y, -y])
        titik_y.extend([y,  y, -y, -y,  x,  x, -x, -x])
        
        if p<0:
            p=p+2*x+3
        else:
            y=y-1
            p=p+2*(x-y)+5
        
        x=x+1

    return titik_x, titik_y

r=7
x_points, y_points = gambar_lingkaran_midpoint(r)

print(f"Lingkaran dengan jari-jari {r}")
print(f"Jumlah titik: {len(x_points)}")
print("10 titik pertama:", list(zip(x_points[:10], y_points[:10])))

plt.figure(figsize=(8,8))
plt.scatter(x_points, y_points, c='blue', s=50)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.5)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.5)
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.title(f"Lingkaran Midpoint(r={r})")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()





        
        
```
