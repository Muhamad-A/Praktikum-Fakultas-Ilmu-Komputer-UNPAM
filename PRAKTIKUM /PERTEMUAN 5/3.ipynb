```python
import matplotlib.pyplot as plt
def gambar_lingkaran_bresenham(r):

    titik_x=[]
    titik_y=[]
    
    x=0
    y=r
    p=3-2*r

    while x<=y:

        titik_x.extend([x, -x, x, -x, y, -y, y, -y])
        titik_y.extend([y, y, -y, -y, x, x, -x, -x])

        x=x+1

        if p<0:
            p = p + 4 * x + 6
        else:
            y = y - 1
            p = p + 4 * (x-y) + 10
        
    return titik_x, titik_y

r=7 
x_points, y_points = gambar_lingkaran_bresenham(r)


print(f"Lingkaran dengan jari-jari {r}")
print(f"Jumlah titik: {len(x_points)}")
print("10 titik pertama:", list(zip(x_points[:10], y_points[:10])))


plt.figure(figsize=(8,8))
plt.scatter(x_points, y_points, c='red', s=50)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.2)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.2)
plt.xlim(-r-2, r+2)
plt.ylim(-r, r+2)
plt.gca().set_aspect('equal')
plt.title(f'Lingkaran Bresenham(r={r})')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

    




    
```
